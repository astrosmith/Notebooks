```python
import yt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline
```

## All data simple colorbar

```python
ds = yt.load('IsolatedGalaxy/galaxy0030/galaxy0030')
ad = ds.all_data()
x = ad['particle_position_x'].in_units('kpc')
y = ad['particle_position_y'].in_units('kpc')
```

```python
print 'Number of particles', len(x)
xmin, xmax = np.min(x), np.max(x)
ymin, ymax = np.min(y), np.max(y)
small_x = 1e-9 * (xmax-xmin)
small_y = 1e-9 * (ymax-ymin)
xmin -= small_x
xmax += small_x
ymin -= small_y
ymax += small_y
print 'min(x) =', xmin.in_units('kpc')
print 'max(x) =', xmax.in_units('kpc')
print 'min(y) =', ymin.in_units('kpc')
print 'max(y) =', ymax.in_units('kpc')
```

```python
bins = 128
cmin = 3
cmap = 'bds_highcontrast'

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
print 'Binning particles. [ bins =', bins, ']'
print 'If a bin has less than', cmin, 'particles we overplot these particles as points.'
delta_x, delta_y = xmax - xmin, ymax - ymin
counts, xedges, yedges, image = plt.hist2d(x, y, bins=bins, range=[[xmin,xmax], [ymin,ymax]], 
                                           cmin=cmin, norm=LogNorm(), cmap=cmap)
plt.colorbar()

mask = np.isnan(counts)
N = len(x) - np.sum(counts.astype(int)[~mask])
if N > 0:
    print 'Overplotting', N, 'particles.'
    ix = ( (bins) * (x - xmin) / (xmax - xmin) ).d.astype(int)
    iy = ( (bins) * (y - ymin) / (ymax - ymin) ).d.astype(int)
    px = yt.YTArray(np.zeros(N), x.units)
    py = yt.YTArray(np.zeros(N), y.units)
    j = 0
    for i in xrange(len(x)):
        if mask[ix[i],iy[i]]:
            px[j] = x[i]
            py[j] = y[i]
            j += 1
    plt.scatter(px, py, c=np.ones(N), #c=iCvec,
           cmap=cmap, norm=image.norm, s=20, edgecolor='none', alpha=.75)

ax.set_xlabel(r'$x\;({\rm kpc})$', fontsize=16)
ax.set_ylabel(r'$y\;({\rm kpc})$', fontsize=16)

plt.show()
```

## The colorbar makes much more sense if we plot (weighted) average quantities

```python
ds = yt.load('IsolatedGalaxy/galaxy0030/galaxy0030')
my_galaxy = ds.disk(ds.domain_center, [0.0, 0.0, 1.0], 10*yt.units.kpc, 3*yt.units.kpc)
x = my_galaxy['particle_position_x'].in_units('kpc')
y = my_galaxy['particle_position_y'].in_units('kpc')
z_string = r'Angular\,momentum'
z = my_galaxy['particle_angular_momentum_magnitude'].in_units('kpc**2*Msun/Myr')
```

```python
print 'Number of particles', len(x)
xmin, xmax = np.min(x), np.max(x)
ymin, ymax = np.min(y), np.max(y)
zmin, zmax = np.min(z), np.max(z)
small_x = 1e-9 * (xmax-xmin)
small_y = 1e-9 * (ymax-ymin)
xmin -= small_x
xmax += small_x
ymin -= small_y
ymax += small_y
print 'min(x) =', xmin
print 'max(x) =', xmax
print 'min(y) =', ymin
print 'max(y) =', ymax
print 'min(z) =', zmin
print 'max(z) =', zmax
```

```python
bins = 128
cmin = 4
cmap = 'Blues'
alpha = .25
interpolation = 'bicubic'

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
print 'Binning particles. [ bins =', bins, ']'
print 'If a bin has less than', cmin, 'particles we overplot these particles as points.'
ix = ( (bins) * (x - xmin) / (xmax - xmin) ).d.astype(int)
iy = ( (bins) * (y - ymin) / (ymax - ymin) ).d.astype(int)
z_no_units = z.d
Z = np.zeros([bins,bins])
W = np.zeros([bins,bins], dtype=int)
for i in xrange(len(x)):
    Z[ix[i],iy[i]] += z_no_units[i]
    W[ix[i],iy[i]] += 1
mask = W <= cmin
Z[~mask] /= W[~mask] # This takes the average
image = plt.imshow(Z.T, origin='lower', extent=[xmin,xmax,ymin,ymax], aspect='auto',
                   interpolation=interpolation, norm=LogNorm(), cmap=cmap)
cbar = plt.colorbar()

N = len(x) - np.sum(W[~mask])
if N > 0:
    print 'Overplotting', N, 'particles.'
    px = yt.YTArray(np.zeros(N), x.units)
    py = yt.YTArray(np.zeros(N), y.units)
    pz = yt.YTArray(np.zeros(N), z.units)
    j = 0
    for i in xrange(len(x)):
        if mask[ix[i],iy[i]]:
            px[j] = x[i]
            py[j] = y[i]
            pz[j] = z[i]
            j += 1
    plt.scatter(px, py, c=pz, cmap=cmap, norm=image.norm, s=20, edgecolor='none', alpha=alpha)

ax.set_xlabel(r'$x\;({\rm %s})$' % x.units, fontsize=16)
ax.set_ylabel(r'$y\;({\rm %s})$' % y.units, fontsize=16)
cbar.ax.set_ylabel(r'${\rm Average\,%s\;(%s)}$' % (z_string, z.units.latex_representation()), fontsize=15)

plt.show()
```
