```python
import yt
```

```python
ds = yt.load("IsolatedGalaxy/galaxy0030/galaxy0030")
ad = ds.all_data()
```

```python
plot = yt.PhasePlot(ad, "density", "temperature", "cell_mass", weight_field=None, fractional=True )
zStr = r'${\rm Gas\,phase:}\;z_{\,}=_{\,}%.1f$' % ds.current_redshift
plot.set_zlim('cell_mass', 1e-9, 1e-1)
plot['cell_mass'].axes.set_title(zStr, fontsize=20)      # using the pyplot API
plot.annotate_text(1e-31, 1e7, zStr, **{'fontsize':20})  # using the yt API
plot.show()
```
